# Introdução e Definição do Problema

> No cenário da nutrição clínica de precisão, o grande desafio não está na escassez de dados, mas na complexidade do cruzamento deles em tempo real. Pacientes acometidos por comorbidades crônicas e gastrointestinais concomitantes exigem uma atenção redobrada no manejo dietético \cite{koudlanouma2023}. Este trabalho foca especificamente no universo de pacientes que sofrem, simultaneamente, de Diabetes Mellitus Tipo 2 (DM2) e Doença do Refluxo Gastroesofágico (DRGE).

> Atualmente, o processo de triagem e montagem de cardápios na clínica modelo é majoritariamente manual ou apoiado em softwares de prontuário estáticos (que apenas listam alimentos, sem inteligência cruzada). O fluxo atual apresenta três dores latentes:

> 1.   Risco de Conflito de Restrições: O nutricionista precisa consultar mentalmente ou em tabelas separadas as diretrizes para ambas as doenças. Por exemplo, o suco de laranja integral possui baixo teor de gordura (neutro para refluxo), mas tem alto índice glicêmico (péssimo para o diabético). Já o chocolate amargo pode ter um índice glicêmico aceitável em pequenas porções, mas é um forte gatilho para o refluxo por relaxar o esfíncter esofágico inferior.

> 2.  Tempo Elevado de Atendimento: O profissional gasta, em média, de 25 a 35 minutos da consulta apenas folheando guias e cruzando tabelas de composição de alimentos para garantir que a prescrição é segura. Esse tempo reduz o espaço para a escuta empática e a educação nutricional do paciente.

> 3.  Falta de Padronização: Diferentes profissionais da mesma clínica podem dar orientações divergentes com base em critérios subjetivos, o que enfraquece a governança clínica e a marca do negócio.

> O problema central pode ser mapeado como uma falha na consistência lógica e checagem de fatos no momento da tomada de decisão. Dispomos de um banco de dados corporativo composto por:

> *  Tabela de Alimentos: Contendo macronutrientes, índice glicêmico, acidez e compostos relaxantes do esfíncter.

> *  Perfil do Paciente: Dados clínicos extraídos do prontuário eletrônico (diagnósticos ativos).

> Nesse sentido, apresentamos nesse trabalho uma solução baseada em IA que atua como um motor de inferência. O sistema recebe o perfil de comorbidades do paciente e a lista de alimentos propostos pelo nutricionista, validando de forma automatizada e instantânea se há alguma violação de diretrizes de saúde. A meta corporativa é reduzir o tempo de validação para menos de 2 segundos, garantindo zero erro de prescrição por conflito de patologias.


# Metodologia e Escolha da Técnica de IA

> Para solucionar o problema de conflito de restrições alimentares em pacientes com Diabetes e Refluxo (DRGE), a metodologia deste trabalho afasta-se de abordagens estatísticas ou probabilísticas tradicionais. Como estamos lidando com a saúde humana, não podemos tolerar margens de erro ou "alucinações" comuns em modelos de aprendizado de máquina puramente indutivos. Necessitamos de uma solução determinística, baseada em regras explícitas e auditoria completa.

> Por essa razão, a técnica de IA selecionada foi a Lógica Proposicional aplicada a Sistemas Baseados em Conhecimento, utilizando o método de Inferência por Resolução para garantir a consistência das dietas \cite{russell2022}.

> A escolha se justifica porque as diretrizes médicas de Nutrição funcionam essencialmente como um sistema axiomático (um conjunto de regras lógicas rígidas). A inferência por resolução nos permite provar, por contradição, se a inclusão de um determinado alimento na dieta do paciente viola alguma restrição de sua condição clínica.

> **Formalização do Modelo Lógico**

> Para estruturar nossa Base de Conhecimento ($BC$), mapeamos as entidades do consultório em proposições lógicas:

> **1. Variáveis Proposicionais:**
> * $D$: O paciente tem Diabetes Tipo 2.
> * $R$: O paciente tem Refluxo (DRGE).
> * $A$: O alimento possui Alto Índice Glicêmico.
> * $C$: O alimento é Ácido ou relaxa o esfíncter esofágico.
> * $P$: O alimento é Proibido para o paciente.

> **2. Regras Médicas da Base de Conhecimento ($BC$):**
> * **Regra 1 (Diabetes):** Se o paciente tem Diabetes e o alimento tem Alto Índice Glicêmico, o alimento é Proibido.
$$D \land A \implies P$$

> * **Regra 2 (Refluxo):** Se o paciente tem Refluxo e o alimento é Ácido/Gatilho, o alimento é Proibido.
$$R \land C \implies P$$

> **3. Mecanismo de Inferência (Resolução):**
> Para automatizar o sistema, transformamos as regras da $BC$ para a Forma Normal Conjuntiva (CNF), que permite a aplicação do algoritmo de resolução computacional:
> * Regra 1 em CNF: $\neg D \lor \neg A \lor P$
> * Regra 2 em CNF: $\neg R \lor \neg C \lor P$

> Quando o nutricionista cadastrar o prontuário de um paciente (ex: informando que ele tem ambas as doenças: $D \land R$) e tentar prescrever suco de laranja integral (que sabemos que possui alto índice glicêmico e acidez: $A \land C$), o motor de inferência adicionará o fato à base e tentará provar a negação do objetivo ($\neg P$).

> Se o algoritmo encontrar uma cláusula vazia ($\square$), fica provado por contradição que o alimento $P$ (Proibido) é uma verdade lógica irrefutável, emitindo um alerta instantâneo de bloqueio no sistema da clínica.

# Implementação da Solução

> Para realizar a validação dietética de maneira robusta, a solução foi codificada utilizando o algoritmo de Inferência por Resolução. Diferente de abordagens baseadas em tabelas de verdade completas (que exigem testar todas as linhas possíveis), o método da resolução foca na manipulação direta das regras lógicas simplificadas. A arquitetura do código está estruturada em duas etapas fundamentais: o Pipeline de Transformação (CNF) e o Loop de Prova por Contradição.

> O algoritmo de resolução exige que todas as sentenças estejam na Forma Normal Conjuntiva (CNF), ou seja, uma grande conjunção de disjunções (um "E" ligando vários blocos de "OU"). O código implementa esse pipeline por meio de funções recursivas que tratam as regras corporativas da clínica:

> 1. `to_cnf(sentence)`: Esta função aplica sistematicamente as equivalências lógicas para normalizar as regras. Ela elimina bicondicionais e implicações ($A \implies B$ vira $\neg A \lor B$), aplica as Leis de De Morgan para mover negações para o interior dos parênteses e faz a distribuição de disjunções ($A \lor (B \land C)$ vira $(A \lor B) \land (A \lor C)$).

> 2. `to_cnf_clauses(sentence)`: Uma vez que a sentença está em CNF, esta função atua como um extrator. Ela quebra a estrutura puramente lógica e a transforma em uma estrutura de dados altamente eficiente para o Python: uma lista de conjuntos (set) de literais. Por exemplo, a regra de restrição do diabético passa a ser representada internamente de forma compacta como {'¬diabetes', '¬alto_indice_glicemico', 'proibido'}.

> O mecanismo de tomada de decisão é controlado pela função `inference_by_resolution(knowledge, query)`. O algoritmo foi parametrizado seguindo os fundamentos da inferência lógica por refutação:

> * **Estado Inicial:** A base de conhecimento da clínica (knowledge) é unida à negação da consulta (Not(query)). Se o nutricionista quer saber se um alimento é de fato proibido ($P$), o algoritmo assume temporariamente que ele não é proibido ($\neg P$) para tentar derrubar essa hipótese.

> * **Mecanismo de Ação (Resolução de Cláusulas):** Utilizando a biblioteca itertools.combinations, o algoritmo emparelha repetidamente todas as cláusulas disponíveis. A função interna resolve(ci, cj) varre o par em busca de literais complementares (como $X$ e $\neg X$). Se encontrados, eles se anulam e uma nova cláusula simplificada é gerada. Tautologias (frases sempre verdadeiras como $X \lor \neg X$) são descartadas por meio da função is_tautology para otimizar o espaço de busca.

> * **Objetivo (Critério de Parada):** O laço de repetição (while True) busca alcançar a cláusula vazia (set()), que representa matematicamente uma contradição insanável ($\square$). Se o set() for gerado, significa que assumir que o alimento não era proibido quebrou as regras da clínica, retornando True (o alimento é proibido). Se o algoritmo rodar, exaurir todas as combinações possíveis e não gerar nenhuma cláusula nova (new.issubset), o loop encerra e retorna False (não há perigo na prescrição).

> O script contendo todo o pipeline de conversão e o motor de resolução encontra-se implementado e pronto para execução na seção de Apêndice.

```python
%%capture
%%writefile logica_proposicional.py
# =====================================================================
# APÊNDICE A: Script de Implementação de Símbolos e Conectores Lógicos
# =====================================================================

import itertools


class Sentence():

    def evaluate(self, model):
        """Evaluates the logical sentence."""
        raise Exception("nothing to evaluate")

    def formula(self):
        """Returns string formula representing logical sentence."""
        return ""

    def symbols(self):
        """Returns a set of all symbols in the logical sentence."""
        return set()

    @classmethod
    def validate(cls, sentence):
        if not isinstance(sentence, Sentence):
            raise TypeError("must be a logical sentence")

    @classmethod
    def parenthesize(cls, s):
        """Parenthesizes an expression if not already parenthesized."""
        def balanced(s):
            """Checks if a string has balanced parentheses."""
            count = 0
            for c in s:
                if c == "(":
                    count += 1
                elif c == ")":
                    if count <= 0:
                        return False
                    count -= 1
            return count == 0
        if not len(s) or s.isalpha() or (
            s[0] == "(" and s[-1] == ")" and balanced(s[1:-1])
        ):
            return s
        else:
            return f"({s})"


class Symbol(Sentence):

    def __init__(self, name):
        self.name = name

    def __eq__(self, other):
        return isinstance(other, Symbol) and self.name == other.name

    def __hash__(self):
        return hash(("symbol", self.name))

    def __repr__(self):
        return self.name

    def evaluate(self, model):
        try:
            return bool(model[self.name])
        except KeyError:
            raise EvaluationException(f"variable {self.name} not in model")

    def formula(self):
        return self.name

    def symbols(self):
        return {self.name}


class Not(Sentence):
    def __init__(self, operand):
        Sentence.validate(operand)
        self.operand = operand

    def __eq__(self, other):
        return isinstance(other, Not) and self.operand == other.operand

    def __hash__(self):
        return hash(("not", hash(self.operand)))

    def __repr__(self):
        return f"Not({self.operand})"

    def evaluate(self, model):
        return not self.operand.evaluate(model)

    def formula(self):
        return "¬" + Sentence.parenthesize(self.operand.formula())

    def symbols(self):
        return self.operand.symbols()


class And(Sentence):
    def __init__(self, *conjuncts):
        for conjunct in conjuncts:
            Sentence.validate(conjunct)
        self.conjuncts = list(conjuncts)

    def __eq__(self, other):
        return isinstance(other, And) and self.conjuncts == other.conjuncts

    def __hash__(self):
        return hash(
            ("and", tuple(hash(conjunct) for conjunct in self.conjuncts))
        )

    def __repr__(self):
        conjunctions = ", ".join(
            [str(conjunct) for conjunct in self.conjuncts]
        )
        return f"And({conjunctions})"

    def add(self, conjunct):
        Sentence.validate(conjunct)
        self.conjuncts.append(conjunct)

    def evaluate(self, model):
        return all(conjunct.evaluate(model) for conjunct in self.conjuncts)

    def formula(self):
        if len(self.conjuncts) == 1:
            return self.conjuncts[0].formula()
        return " ∧ ".join([Sentence.parenthesize(conjunct.formula())
                           for conjunct in self.conjuncts])

    def symbols(self):
        return set.union(*[conjunct.symbols() for conjunct in self.conjuncts])


class Or(Sentence):
    def __init__(self, *disjuncts):
        for disjunct in disjuncts:
            Sentence.validate(disjunct)
        self.disjuncts = list(disjuncts)

    def __eq__(self, other):
        return isinstance(other, Or) and self.disjuncts == other.disjuncts

    def __hash__(self):
        return hash(
            ("or", tuple(hash(disjunct) for disjunct in self.disjuncts))
        )

    def __repr__(self):
        disjuncts = ", ".join([str(disjunct) for disjunct in self.disjuncts])
        return f"Or({disjuncts})"

    def evaluate(self, model):
        return any(disjunct.evaluate(model) for disjunct in self.disjuncts)

    def formula(self):
        if len(self.disjuncts) == 1:
            return self.disjuncts[0].formula()
        return " ∨  ".join([Sentence.parenthesize(disjunct.formula())
                            for disjunct in self.disjuncts])

    def symbols(self):
        return set.union(*[disjunct.symbols() for disjunct in self.disjuncts])


class Implication(Sentence):
    def __init__(self, antecedent, consequent):
        Sentence.validate(antecedent)
        Sentence.validate(consequent)
        self.antecedent = antecedent
        self.consequent = consequent

    def __eq__(self, other):
        return (isinstance(other, Implication)
                and self.antecedent == other.antecedent
                and self.consequent == other.consequent)

    def __hash__(self):
        return hash(("implies", hash(self.antecedent), hash(self.consequent)))

    def __repr__(self):
        return f"Implication({self.antecedent}, {self.consequent})"

    def evaluate(self, model):
        return ((not self.antecedent.evaluate(model))
                or self.consequent.evaluate(model))

    def formula(self):
        antecedent = Sentence.parenthesize(self.antecedent.formula())
        consequent = Sentence.parenthesize(self.consequent.formula())
        return f"{antecedent} => {consequent}"

    def symbols(self):
        return set.union(self.antecedent.symbols(), self.consequent.symbols())


class Biconditional(Sentence):
    def __init__(self, left, right):
        Sentence.validate(left)
        Sentence.validate(right)
        self.left = left
        self.right = right

    def __eq__(self, other):
        return (isinstance(other, Biconditional)
                and self.left == other.left
                and self.right == other.right)

    def __hash__(self):
        return hash(("biconditional", hash(self.left), hash(self.right)))

    def __repr__(self):
        return f"Biconditional({self.left}, {self.right})"

    def evaluate(self, model):
        return ((self.left.evaluate(model)
                 and self.right.evaluate(model))
                or (not self.left.evaluate(model)
                    and not self.right.evaluate(model)))

    def formula(self):
        left = Sentence.parenthesize(str(self.left))
        right = Sentence.parenthesize(str(self.right))
        return f"{left} <=> {right}"

    def symbols(self):
        return set.union(self.left.symbols(), self.right.symbols())
```

```python
%%capture
%%writefile inferencia_por_resolucao.py
# =====================================================================
# APÊNDICE B: Script de Implementação da Inferência por Resolução
# =====================================================================

import itertools

def inference_by_resolution(knowledge, query):
    """
    Determina se a query é uma consequência lógica da base de conhecimento
    usando o algoritmo de resolução.
    """

    # 1. KB contém (knowledge AND NOT query)
    # Vamos trabalhar com uma lista de cláusulas.
    # Cada cláusula é um set de literais.
    clauses = []

    # Adicionamos o conhecimento e a negação da query
    sentences = [knowledge, Not(query)]

    # 2. Converter sentenças para CNF e extrair cláusulas
    for sentence in sentences:
        clauses.extend(to_cnf_clauses(sentence))

    # Remover duplicatas dentro das cláusulas e cláusulas redundantes
    clauses = [set(c) for c in clauses]

    new = set()

    while True:
        # Gerar todos os pares possíveis de cláusulas para tentar resolver
        n = len(clauses)
        pairs = list(itertools.combinations(range(n), 2))

        for (i, j) in pairs:
            # Tenta resolver o par de cláusulas
            resolvents = resolve(clauses[i], clauses[j])

            # Se encontrar a cláusula vazia (set vazio), a contradição foi provada
            if set() in resolvents:
                return True

            # Adiciona novas cláusulas resultantes ao conjunto 'new'
            for res in resolvents:
                new.add(frozenset(res))

        # Se não houver novas cláusulas para adicionar, não há contradição
        if new.issubset(set(frozenset(c) for c in clauses)):
            return False

        # Adiciona as novas cláusulas à lista principal
        for n_clause in new:
            if set(n_clause) not in clauses:
                clauses.append(set(n_clause))

def resolve(ci, cj):
    """Retorna o conjunto de todas as cláusulas possíveis resultantes da resolução de ci e cj."""
    resolvents = []
    for di in ci:
        for dj in cj:
            # Se um é a negação do outro (ex: P e Not(P))
            if di == Not(dj) or Not(di) == dj:
                # Cria uma nova cláusula removendo os literais complementares
                new_clause = (ci - {di}) | (cj - {dj})
                # Otimização: se a nova cláusula for uma tautologia (P ou Not P), ignoramos
                if not is_tautology(new_clause):
                    resolvents.append(new_clause)
    return resolvents

def is_tautology(clause):
    """Verifica se uma cláusula é sempre verdadeira (ex: possui P e Not(P))."""
    for lit in clause:
        if Not(lit) in clause:
            return True
    return False

def to_cnf(sentence):
    """Converte uma sentença lógica para a Forma Normal Conjuntiva (CNF)."""

    # 1. Eliminar Bicondicionais (A <=> B vira (A => B) & (B => A))
    if isinstance(sentence, Biconditional):
        return to_cnf(And(Implication(sentence.left, sentence.right),
                          Implication(sentence.right, sentence.left)))

    # 2. Eliminar Implicações (A => B vira Not(A) Or B)
    if isinstance(sentence, Implication):
        return to_cnf(Or(Not(sentence.antecedent), sentence.consequent))

    # 3. Mover Negações para dentro (De Morgan)
    if isinstance(sentence, Not):
        op = sentence.operand
        if isinstance(op, Symbol):
            return sentence
        if isinstance(op, Not): # Dupla negação
            return to_cnf(op.operand)
        if isinstance(op, And): # Not(A and B) -> (Not A or Not B)
            return to_cnf(Or(*[Not(c) for c in op.conjuncts]))
        if isinstance(op, Or):  # Not(A or B) -> (Not A and Not B)
            return to_cnf(And(*[Not(d) for d in op.disjuncts]))
        return sentence

    # 4. Distribuir OU sobre E: A Or (B And C) -> (A Or B) And (A Or C)
    if isinstance(sentence, Or):
        # Primeiro, garante que os filhos já estão em CNF
        disjuncts = [to_cnf(d) for d in sentence.disjuncts]

        # Procura por um componente que seja um 'And' para distribuir
        for i, d in enumerate(disjuncts):
            if isinstance(d, And):
                others = disjuncts[:i] + disjuncts[i+1:]
                return to_cnf(And(*[Or(c, *others) for c in d.conjuncts]))
        return Or(*disjuncts)

    # 5. Processar conjunções recursivamente
    if isinstance(sentence, And):
        return And(*[to_cnf(c) for c in sentence.conjuncts])

    return sentence

def to_cnf_clauses(sentence):
    """
    Transforma a sentença em CNF e depois quebra em uma lista de conjuntos.
    Ex: (A v B) ^ (C v D) -> [{A, B}, {C, D}]
    """
    cnf_sentence = to_cnf(sentence)
    clauses = []

    def extract_disjuncts(s):
        if isinstance(s, Or):
            result = set()
            for d in s.disjuncts:
                result.update(extract_disjuncts(d))
            return result
        else:
            return {s}

    def extract_conjuncts(s):
        if isinstance(s, And):
            for c in s.conjuncts:
                extract_conjuncts(c)
        else:
            clauses.append(extract_disjuncts(s))

    extract_conjuncts(cnf_sentence)
    return clauses
```

> * CENÁRIO 2: Paciente apenas Diabético / Melancia (Alto IG, mas Não Ácida) / Bloqueado
> * CENÁRIO 3: Paciente apenas com Refluxo / Chocolate Amargo (Gatilho, mas Baixo IG) / Bloqueado
> * CENÁRIO 4: Paciente com ambas as doenças / Brócolis Cozido (Seguro para ambos) / Liberado


```python
%%capture
%%writefile simbolos_consultorio.py
# =====================================================================
# APÊNDICE C: Script de Implementação dos Símbolos do Consulório
# =====================================================================

D = Symbol("Paciente tem Diabetes Tipo 2")
R = Symbol("Paciente tem Refluxo (DRGE)")

A = Symbol("Alimento possui Alto Índice Glicêmico")
C = Symbol("Alimento é Ácido / Gatilho de Refluxo")

P = Symbol("Alimento é Proibido para o Paciente")

# ==========================================
# BASE DE REGRAS MÉDICAS DA CLÍNICA (Regras Estáticas)
# ==========================================
Regras_Clinicas = And(
    Implication(And(D, A), P),  # Se Diabético e Alto IG -> Proibido
    Implication(And(R, C), P)   # Se Refluxo e Ácido -> Proibido
)

print("--- EXECUTANDO TRIAGEM DIETÉTICA AUTOMATIZADA ---\n")

# -------------------------------------------------------------
# CENÁRIO 1: Paciente com ambas as doenças / Suco de Laranja (Alto IG e Ácido)
# -------------------------------------------------------------
KB_cenario_1 = And(
    Regras_Clinicas,
    D,          # Prontuário: Tem Diabetes
    R,          # Prontuário: Tem Refluxo
    A,          # Alimento: Suco de Laranja tem Alto IG
    C           # Alimento: Suco de Laranja é Ácido
)
res_1 = inference_by_resolution(KB_cenario_1, P)
print(f"Cenário 1 (Suco de Laranja para Paciente D + R) -> Alimento Proibido? {res_1}")


# -------------------------------------------------------------
# CENÁRIO 2: Paciente apenas Diabético / Melancia (Alto IG, mas Não Ácida)
# -------------------------------------------------------------
KB_cenario_2 = And(
    Regras_Clinicas,
    D,          # Prontuário: Tem Diabetes
    Not(R),     # Prontuário: NÃO tem Refluxo
    A,          # Alimento: Melancia tem Alto IG
    Not(C)      # Alimento: Melancia NÃO é ácida
)
res_2 = inference_by_resolution(KB_cenario_2, P)
print(f"Cenário 2 (Melancia para Paciente Apenas Diabético) -> Alimento Proibido? {res_2}")


# -------------------------------------------------------------
# CENÁRIO 3: Paciente apenas com Refluxo / Chocolate Amargo (Gatilho, mas Baixo IG)
# -------------------------------------------------------------
KB_cenario_3 = And(
    Regras_Clinicas,
    Not(D),     # Prontuário: NÃO tem Diabetes
    R,          # Prontuário: Tem Refluxo
    Not(A),     # Alimento: Chocolate Amargo tem baixo IG
    C           # Alimento: Chocolate Amargo relaxa o esfíncter (Gatilho)
)
res_3 = inference_by_resolution(KB_cenario_3, P)
print(f"Cenário 3 (Chocolate para Paciente Apenas com Refluxo) -> Alimento Proibido? {res_3}")


# -------------------------------------------------------------
# CENÁRIO 4: Paciente com ambas as doenças / Brócolis Cozido (Seguro para ambos)
# -------------------------------------------------------------
KB_cenario_4 = And(
    Regras_Clinicas,
    D,          # Prontuário: Tem Diabetes
    R,          # Prontuário: Tem Refluxo
    Not(A),     # Alimento: Brócolis NÃO tem Alto IG
    Not(C)      # Alimento: Brócolis NÃO é ácido nem gatilho
)
res_4 = inference_by_resolution(KB_cenario_4, P)
print(f"Cenário 4 (Brócolis para Paciente D + R) -> Alimento Proibido? {res_4}")
```

## Avaliação Comparativa de Larga Escala: Humano vs. Inteligência Artificial

> Para mensurar o Retorno sobre o Investimento (ROI) e o ganho de segurança clínica da solução, foi desenvolvida uma simulação contendo 100 cenários de pacientes com sintomas variados e diferentes propostas de alimentos.

### Estruturação dos Dados de Simulação

> Os dados foram consolidados em uma lista de dicionários Python. Esta estrutura mimetiza o histórico de atendimentos reais da clínica e armazena os seguintes atributos corporativos:

> * **Prontuário e Atributos do Alimento:** Variáveis booleanas que alimentam a Base de Conhecimento.

> * **Métricas Humanas (Baseline):** O tempo médio (em segundos) que um nutricionista leva para folhear guias tradicionais e sua respectiva decisão (sujeita a fadiga e erros).

> * **Ground Truth:** A verdade absoluta baseada estritamente nas diretrizes médicas consolidadas (padrão ouro).

```python
%%capture
%%writefile avaliacao_larga_escala.py
# =====================================================================
# APÊNDICE D: Script de Implementação de Avaliação em Larga Escala
# =====================================================================

import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Configuração de semente para reproducibilidade da simulação
np.random.seed(42)

# =====================================================================
# 1. ESTRUTURAÇÃO DA BASE DE DADOS COMPREENSIVA (100 Cenários)
# =====================================================================
cenarios_corporativos = []

for i in range(100):
    # Geração aleatória de perfis clínicos e propriedades alimentares
    tem_diabetes = np.random.choice([True, False])
    tem_refluxo = np.random.choice([True, False])
    alto_ig = np.random.choice([True, False])
    eh_acido = np.random.choice([True, False])

    # Ground Truth: Regra de IA pura determinada pelas diretrizes médicas
    deve_proibir = (tem_diabetes and alto_ig) or (tem_refluxo and eh_acido)

    # Simulação do Comportamento Humano (Profissional consultando tabelas manuais)
    # Tempo humano varia de 15 a 45 segundos por análise de alimento
    tempo_humano = np.random.uniform(15.0, 45.0)

    # Taxa de erro humana estimada em ~15% devido à fadiga e falta de padronização
    if np.random.rand() < 0.15:
        decisao_humana = not deve_proibir  # Humano errou o diagnóstico
    else:
        decisao_humana = deve_proibir      # Humano acertou o diagnóstico

    cenarios_corporativos.append({
        "id": i + 1,
        "features_paciente": (tem_diabetes, tem_refluxo),
        "features_alimento": (alto_ig, eh_acido),
        "tempo_humano": tempo_humano,
        "decisao_humana": decisao_humana,
        "ground_truth": deve_proibir
    })

# =====================================================================
# 2. EXECUÇÃO DO PROCESSO DE AUDITORIA PELA INTELIGÊNCIA ARTIFICIAL
# =====================================================================
predicoes_ia = []
tempos_ia = []

# Loop automatizado simulando a varredura da IA sobre a base corporativa
for cenario in cenarios_corporativos:
    # Captura dos dados estruturados
    d_pt, r_pt = cenario["features_paciente"]
    a_al, c_al = cenario["features_alimento"]

    # Instanciação dinâmica da KB com a sintaxe das nossas classes lógicas
    # (Considerando as classes implementadas na Seção 4)
    KB_dinamica = And(
        Regras_Clinicas,
        Symbol("Paciente tem Diabetes Tipo 2") if d_pt else Not(Symbol("Paciente tem Diabetes Tipo 2")),
        Symbol("Paciente tem Refluxo (DRGE)") if r_pt else Not(Symbol("Paciente tem Refluxo (DRGE)")),
        Symbol("Alimento possui Alto Índice Glicêmico") if a_al else Not(Symbol("Alimento possui Alto Índice Glicêmico")),
        Symbol("Alimento é Ácido / Gatilho de Refluxo") if c_al else Not(Symbol("Alimento é Ácido / Gatilho de Refluxo"))
    )

    # Medição de tempo de altíssima precisão para a resposta da IA
    t_inicio = time.perf_counter()
    resposta_ia = inference_by_resolution(KB_dinamica, Symbol("Alimento é Proibido para o Paciente"))
    t_fim = time.perf_counter()

    predicoes_ia.append(resposta_ia)
    tempos_ia.append(t_fim - t_inicio)

# Consolidação das respostas para análise estatística
y_true = [c["ground_truth"] for c in cenarios_corporativos]
y_humano = [c["decisao_humana"] for c in cenarios_corporativos]
y_ia = predicoes_ia

tempos_humanos_totais = [c["tempo_humano"] for c in cenarios_corporativos]
```

## Organização e Engenharia dos Dados de Entrada ($X$)

>> Para que uma rede neural profunda processe as informações do atendimento, precisamos transformar os dados brutos e heterogêneos da clínica em um tensor numérico padronizado:

>> * **Dados do Paciente (Perfil Clínico):** O prontuário médico extraído do sistema SQL seria transformado em um vetor binário através de Multi-Hot Encoding, onde cada posição representa a presença (1) ou ausência (0) de uma patologia em um catálogo de, por exemplo, 200 doenças mapeadas pelo CID (Código Internacional de Doenças).
>> * **Dados do Alimento (Propriedades Nutricionais):** Os alimentos seriam representados por variáveis contínuas (contagem de gramas de carboidratos, proteínas, lipídios, fibras, nível de pH e miligramas de sódio). Para evitar que variáveis com escalas maiores (como calorias) dominem o aprendizado do modelo em detrimento do pH, aplicaríamos a Normalização Min-Max ou o Z-score Standardization, travando todos os valores contínuos estritamente no intervalo $[0, 1]$ ou $[-1, 1]$.
>> * **Fusão dos Dados (Vetor $X$):** O vetor de entrada final $X$ seria a concatenação desses dois blocos (Vetor Paciente + Vetor Alimento), servindo como uma assinatura única da relação paciente-alimento.

## Representação e Arquitetura Teórica da Rede

>> Dado que a entrada consolidada $X$ possui um formato tabular (uma matriz de características estruturadas), a arquitetura mais adequada e eficiente é uma Rede Neural Totalmente Conectada (MLP - Multilayer Perceptron) profunda:

>> * **Camadas Ocultas (Hidden Layers):** A rede contaria com uma camada de entrada, seguida por 3 a 4 camadas densas intermediárias (diminuindo progressivamente em número de neurônios, ex: 128, 64, 32) para extrair as correlações complexas e ocultas entre nutrientes e respostas biológicas.
>> * **Funções de Ativação:** Utilizaríamos a função ReLU (Rectified Linear Unit) em todas as camadas ocultas. A escolha da ReLU se justifica por sua capacidade de quebrar a linearidade do modelo — permitindo capturar o fato de que a acidez de um alimento só se torna perigosa se combinada com o relaxamento prévio do esfíncter esofágico —, além de mitigar o problema do Vanishing Gradient durante o contratreinamento.
>> * **Regularização:** Para mitigar o risco de *overfitting* (onde a rede decora o histórico da clínica mas falha ao atender um paciente novo), incluiríamos camadas de Dropout com taxa de 20% entre as camadas densas.

## Estrutura da Saída Esperada ($Y$)

>> O objetivo final da rede é classificar a recomendação do alimento em um espectro clínico acionável para o nutricionista.

>> * **Camada de Saída:** A última camada da rede conteria 3 neurônios, correspondentes a três classes de decisão da clínica: [Altamente Recomendado, Neutro/Moderado, Proibido/Perigoso].
>> * **Função de Ativação Final:** Utilizaríamos a função Softmax. A Softmax pegará os valores brutos de saída da rede (logits) e os transformará em uma distribuição de probabilidade que soma $100\%$ (ex: $92\%$ de chance de ser Proibido, $6\%$ Neutro, $2\%$ Recomendado). Isso traz transparência, pois permite ao nutricionista ver o nível de certeza da IA antes de tomar uma decisão de bloqueio.
>> * **Função de Perda (Loss Function):** A baliza matemática para o treinamento seria a Categorical Cross-Entropy (Entropia Cruzada Categórica). Ela medirá a distância entre as probabilidades previstas pela Softmax e os rótulos reais de prescrição (Ground Truth). O erro calculado por essa função guiará o otimizador Adam a ajustar os pesos da rede através do algoritmo de retropropagação (backpropagation).

>> Com essa modelagem teórica, a clínica substitui um sistema rígido de "Sim ou Não" por uma plataforma preditiva capaz de aprender continuamente com o histórico de sucesso dos tratamentos passados, elevando a escala do software a níveis globais de processamento.

## Casos de Borda (Edge Cases) e Intervenção Humana

> A IA opera sob a premissa do "Mundo Fechado", o que significa que ela é infalível apenas dentro do limite das regras médicas previamente cadastradas. Os principais edge cases (casos de borda) identificados onde a IA pode falhar ou gerar soluções impraticáveis na vida real são:

>> * **Interações Medicamentosas Ocultas:** O sistema pode liberar um alimento que é seguro para Diabetes e Refluxo, mas que corta o efeito de um medicamento de uso contínuo que o paciente toma (ex: o consumo de laranja/uva interferindo em estatinas).

>> * **Condições Clínicas Omissas:** Pacientes grávidas com Diabetes Gestacional ou indivíduos com insuficiência renal crônica associada demandam restrições calóricas e de potássio que o motor atual não computa.

>> **Ação de Mitigação:** A IA foi desenhada para ser um Sitema de Suporte à Decisão (*Human-in-the-loop*) e nunca um substituto autônomo. O veredicto do software é uma recomendação consultiva; o botão de salvar o cardápio exige a assinatura digital e validação soberana do nutricionista responsável.

## Viés (Bias) e Injustiça nos Dados

> Embora a lógica formal seja matematicamente neutra, os dados alimentados nela podem perpetuar vieses corporativos ou socioeconômicos:

>> * **Viés de Classe e Disponibilidade:** Se as regras de substituição de alimentos forem baseadas estritamente em ingredientes caros, importados ou de difícil acesso (ex: prescrever apenas salmão ou farinhas especiais), o algoritmo criará planos alimentares excludentes para pacientes de menor poder aquisitivo, gerando uma barreira de adesão ao tratamento.

>>  **Ação de Mitigação:** A base de conhecimento deve ser continuamente auditada por comitês clínicos para garantir uma representação democrática e diversa de alimentos regionais e acessíveis, garantindo que o algoritmo promova a equidade na saúde.

## Explicabilidade (XAI - Explainable AI)

> Um dos maiores trunfos da abordagem por Lógica Proposicional e Resolução utilizada neste trabalho é a sua explicabilidade nativa. Ao contrário de modelos de caixa-preta (*Black-Box*) de Machine Learning, onde é difícil rastrear o motivo de um neurônio ter tomado uma decisão, o motor de resolução permite fazer engenharia reversa completa de cada decisão.

> Se o sistema bloquear um alimento e o nutricionista ou o paciente questionarem o motivo, o software consegue imprimir a "árvore de refutação", traduzindo-a comercialmente da seguinte forma:

>> *"O Alimento [Suco de Laranja] foi classificado como PROIBIDO porque o paciente possui a condição ativa [Diabetes Tipo 2] e o insumo aciona a regra de segurança médica nº 1, que veda itens com [Alto Índice Glicêmico].".*

> Essa transparência mitiga riscos jurídicos de responsabilidade civil para a clínica, constrói uma relação de extrema confiança com o cliente final e eleva o nível técnico de toda a equipe médica, consolidando o sucesso do projeto.

```python
import json
import os
import re
from datetime import datetime

# 1. Configuração da data atual
meses = {1: "janeiro", 2: "fevereiro", 3: "março", 4: "abril", 5: "maio", 6: "junho",
         7: "julho", 8: "agosto", 9: "setembro", 10: "outubro", 11: "novembro", 12: "dezembro"}
hoje = datetime.now()
data_formatada = f"{hoje.day} de {meses[hoje.month]} de {hoje.year}"

# 2. Carrega o notebook e remove as células de execução técnica para limpar o PDF
with open('/content/Resposta_Trabalho_Trilha-I.ipynb', 'r', encoding='utf-8') as f:
    nb_data = json.load(f)

celulas_filtradas = [c for c in nb_data['cells'] if 'nbconvert' not in "".join(c.get('source', [])) and 'xelatex' not in "".join(c.get('source', [])) and 'referencias.bib' not in "".join(c.get('source', []))]
nb_data['cells'] = celulas_filtradas

with open('/content/temp_limpo.ipynb', 'w', encoding='utf-8') as f:
    json.dump(nb_data, f, ensure_ascii=False, indent=2)

# 3. Converte para LaTeX intermediário
!jupyter nbconvert --to latex /content/temp_limpo.ipynb --no-input --output=Resposta_Trabalho_Trilha-I

# 4. Ajustes finos de pacotes, sumários e capas no arquivo .tex
!sed -i '/\\maketitle/d' /content/Resposta_Trabalho_Trilha-I.tex

# 5. DEFESA CONTRA A TRAVA DO NBCONVERT: Apaga o comando que desativa as legendas
!sed -i '/nocaption/d' /content/Resposta_Trabalho_Trilha-I.tex

# Processamento Estrutural Avançado para Capa + Folha de Rosto
with open('/content/Resposta_Trabalho_Trilha-I.tex', 'r', encoding='utf-8') as f:
    conteudo = f.read()

# Aplica a data dinâmica
if '\\today' in conteudo:
    conteudo = conteudo.replace('\\today', data_formatada)

# Injeta pacotes e traduções globais do LaTeX
if '\\begin{document}' in conteudo:
    traducoes = (
        '\\usepackage[brazil]{babel}\n' # Muda Figure/Table para Figura/Tabela
        '\\usepackage{float}\n'
        '\\usepackage{url}\n'
        '\\usepackage{booktabs}\n' # Pacote necessário para ler tabelas do pandas de forma elegante
        '\\usepackage[aboveskip=6pt,position=top]{caption}\n' # Força legendas de tabelas no topo (ABNT)
        '\\captionsetup[table]{position=top}\n'  # Garante a posição superior para tabelas
        '\\captionsetup[figure]{position=bottom}\n' # Garante a posição inferior para figuras
        '\\usepackage{listings}\n' # Suporte a listagem de código
        '\\usepackage{xcolor}\n'   # Permite colorir a sintaxe do código
        '\\definecolor{codegreen}{rgb}{0,0.6,0}\n'
        '\\definecolor{codegray}{rgb}{0.5,0.5,0.5}\n'
        '\\definecolor{codepurple}{rgb}{0.58,0,0.82}\n'
        '\\definecolor{backcolour}{rgb}{0.95,0.95,0.92}\n'
        '\\lstset{\n'        # Configuração visual do bloco de código Python
        '    backgroundcolor=\\color{backcolour},\n'
        '    commentstyle=\\color{codegreen},\n'
        '    keywordstyle=\\color{magenta},\n'
        '    numberstyle=\\small\\color{codegray},\n'
        '    stringstyle=\\color{codepurple},\n'
        '    basicstyle=\\ttfamily\\small,\n'
        '    breakatwhitespace=false,\n'
        '    breaklines=true,\n'
        '    captionpos=t,\n'
        '    keepspaces=true,\n'
        '    numbers=left,\n'
        '    numbersep=5pt,\n'
        '    showspaces=false,\n'
        '    showstringspaces=false,\n'
        '    showtabs=false,\n'
        '    tabsize=4\n'
        '}\n'
        '\\renewcommand{\\familydefault}{\\sfdefault}\n' # OBRIGA TODO O DOCUMENTO (INCLUINDO A CAPA) A USAR A MESMA FONTE SANS-SERIF
        '\\renewcommand{\\contentsname}{Sumário}\n'
        '\\renewcommand{\\listfigurename}{Lista de Figuras}\n'
        '\\renewcommand{\\listtablename}{Lista de Tabelas}\n' # Lista nova
        '\\renewcommand{\\figurename}{Figura}\n' # FORÇA A TRADUÇÃO DE FIGURE
        '\\renewcommand{\\tablename}{Tabela}\n' # FORÇA A TRADUÇÃO DE TABLE
        '\\renewcommand{\\refname}{Referências}\n'
        '\\begin{document}'
    )
    conteudo = conteudo.replace('\\begin{document}', traducoes)

# Posiciona Sumário e Lista de Figuras após a capa do IF Goiano
# Localiza a SEGUNDA ocorrência de \end{titlepage} para injetar os índices após a folha de rosto
partes_titlepage = conteudo.split('\\end{titlepage}')
if len(partes_titlepage) >= 3:
    # Cria o bloco de índices acadêmicos automáticos
    bloco_preambulo = (
        '\n\\tableofcontents\n\\newpage\n'
        '\\listoffigures\n\\newpage\n'
        '\\listoftables\n\\newpage\n'
    )
    # Reassenta o texto colocando as listas após a folha de rosto (segunda titlepage)
    conteudo = partes_titlepage[0] + '\\end{titlepage}' + partes_titlepage[1] + '\\end{titlepage}\n' + bloco_preambulo + '\\end{titlepage}'.join(partes_titlepage[2:])

# --- ORDENAÇÃO FIXA DA ABNT: REFERÊNCIAS ANTES DO APÊNDICE ---
# Extrai o conteúdo do apêndice do final para reinjetá-lo após as Referências
if '\\appendix' in conteudo:
    partes_apendice = conteudo.split('\\appendix')
    conteudo_principal = partes_apendice[0].replace('\\end{document}', '')
    conteudo_apendice = '\\appendix' + partes_apendice[1].replace('\\end{document}', '')

    # Monta a estrutura correta: Principal -> Referências -> Apêndice -> Fim
    conteudo = (
        conteudo_principal +
        '\n\\newpage\n\\addcontentsline{toc}{section}{Referências}\n\\bibliographystyle{plain}\n\\bibliography{referencias}\n' +
        '\n\\clearpage\n' + conteudo_apendice +
        '\n\\end{document}'
    )
else:
  # --- INJEÇÃO AUTOMÁTICA DO COMANDO BIBTEX NO .TEX ---
  bloco_bibtex = (
      '\n\\newpage\n'
      '\\addcontentsline{toc}{section}{Referências}\n'
      '\\bibliographystyle{plain}\n' # 'plain' organiza por ordem alfabética do autor principal
      '\\bibliography{referencias}\n'
      '\\end{document}'
  )
  if '\\end{document}' in conteudo:
      conteudo = conteudo.replace('\\end{document}', bloco_bibtex)

with open('/content/Resposta_Trabalho_Trilha-I.tex', 'w', encoding='utf-8') as f:
    f.write(conteudo)

# 5. EXECUÇÃO INTEGRAL DO MOTOR BIBTEX DO LINUX
# Remove arquivos auxiliares antigos para evitar lixo de compilações passadas
for ext in ['.aux', '.bbl', '.blg', '.pdf', '.toc', '.lof']:
    if os.path.exists('Resposta_Trabalho_Trilha-I' + ext):
        os.remove('Resposta_Trabalho_Trilha-I' + ext)

print("Compilando 1/3 (Processando referências internas)...")
!xelatex -interaction=batchmode Resposta_Trabalho_Trilha-I.tex > /dev/null
print("Compilando 2/3 (Ligando arquivo referencias.bib)...")
!bibtex Resposta_Trabalho_Trilha-I > /dev/null
print("Compilando 3/3 (Sincronizando índices e sumários)...")
!xelatex -interaction=batchmode Resposta_Trabalho_Trilha-I.tex > /dev/null
!xelatex -interaction=batchmode Resposta_Trabalho_Trilha-I.tex > /dev/null

print("Concluído! Pronto para download.")
```
